```python
df = pd.read_csv("SSA Baby Names - All SSA Data.csv")
```

```python
df.dtypes
```

## Tiana:

```python
tiana_df = df[(df['Name'] == 'Tiana')]
tiana_df = tiana_df[(tiana_df['Sex'] == 'F')]
tiana_00_df = tiana_df[(tiana_df['Year'] > 2000)]
tiana_total = tiana_00_df['Count'].sum()
tiana_total
```

```python
tiana_00_df = tiana_00_df.reset_index(drop=True).drop(columns=['Sex', 'Name'])
tiana_00_df
```

```python
tiana_chart = alt.Chart(tiana_00_df).mark_line().encode(
    x='Year',
    y='Count'
).properties(
    title='Line Graph of tiana_00_df',
    width=600,  
    height=400 )
```
